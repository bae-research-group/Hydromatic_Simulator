```python
# ----------- Encoding layer + LSTM (timestep: -1) -----------
class EncodingModule(keras.Model):

    def __init__(self):
        super().__init__()
        # Design embedder
        self.embed_design = layers.Dense(settings['feature-dim'], activation='relu')
        self.lstm_cell = layers.LSTMCell(settings['lstm-dim'])

    def call(self, inputs, hidden_states, training):
        if hidden_states == None:
            batch_size = tf.shape(inputs)[0]
            hidden_states = tf.zeros([2, batch_size, settings['lstm-dim']], dtype=tf.float32)

        feature_vector = self.embed_design(inputs, training=training)
        _, new_hidden_states = self.lstm_cell(
            inputs=feature_vector, states=hidden_states, training=training)
        return new_hidden_states

# ----------- LSTM (timestep: 0 to T-1) -----------
class DecodingModule(keras.Model):

    def __init__(self):
        super().__init__()
        # Position embedder and LSTM
        self.embed_pos = layers.Dense(settings['feature-dim'], activation='relu')
        self.lstm_cell = layers.LSTMCell(settings['lstm-dim'])

        # Output layers
        self.dense1 = layers.Dense(settings['dense-dim'], activation='relu')
        self.dense2 = layers.Dense(settings['dense-dim'], activation='relu')
        self.dense3 = layers.Dense(settings['nodal-dim'], activation='linear')

    def call(self, inputs, hidden_states, training):

        x = self.embed_pos(inputs, training=training)
        x, new_hidden_states = self.lstm_cell(
            x, states=hidden_states, training=training)

        x = self.dense1(x, training=training)
        x = self.dense2(x, training=training)
        out = self.dense3(x, training=training)
        return out, new_hidden_states
```

```python
# ----------- Helper class that trains encoder/decoder -----------
class ModelTrainer(keras.Model):

    def __init__(self):
        super().__init__()

        self.encoder = EncodingModule()
        self.decoder = DecodingModule()

        self.loss_tracker = keras.metrics.Mean(name='loss')
        self.acc_tracker1 = keras.metrics.Mean(name='mean_ed') # predictions with teacher forcing
        self.acc_tracker2 = keras.metrics.Mean(name='max_ed')
        self.R2_tracker = keras.metrics.Mean(name='R2')


    def call_encoder(self, inputs, training=False):
        return self.encoder(inputs, None, training=training)

    def call_decoder(self, inputs, hidden_states, training=False):
        return self.decoder(inputs, hidden_states, training=training)

    def round_differentiable(self, x, decimal_pt):
      factor = 10.0 ** decimal_pt
      x_scaled = x * factor
      x_nondiff = tf.round(x_scaled) / factor

      return x + tf.stop_gradient(x_nondiff - x) #-- makes rounding differentiable in backprop


    def teacher_forcing_generate(self, design_input, nodal_input_seq, training=False):

        nodal_output_seq = []

        # ----------- Encoding layer + LSTM (timestep: -1) -----------
        new_hidden_states = self.call_encoder(
            design_input, training=training
        )
        # ----------- LSTM (timestep: 0 to T-1) -----------
        for i in range(settings['num-timesteps']):
            nodal_input = nodal_input_seq[:,i,:]
            hidden_states = new_hidden_states

            next_pos, new_hidden_states = self.call_decoder(
                nodal_input, hidden_states, training=training
            )
            nodal_output_seq.append(next_pos)

        nodal_output_seq = tf.transpose(tf.stack(nodal_output_seq), perm=[1,0,2])  # shape: (TOTAL_TIMEPOINT_NO, -1, shape_dim) -> (-1, TOTAL_TIMEPOINT_NO, shape_dim)
        return self.round_differentiable(nodal_output_seq, settings['round-decimal'])


    def recursive_generate(self, design_input, nodal_input, training=False):
        nodal_output_seq = []

        # ----------- Encoding layer + LSTM (timestep: -1) -----------
        new_hidden_states = self.call_encoder(
            design_input, training=training
        )

        # ----------- LSTM (timestep: 0 to T-1) --------------------
        #-------- Input initial nodal point (without Teacher forcing)
        hidden_states = new_hidden_states

        next_pos, new_hidden_states = self.call_decoder(
            nodal_input, hidden_states, training=training
        )
        nodal_output_seq.append(next_pos)

        for i in range(1,settings['num-timesteps']):
            nodal_input = next_pos
            hidden_states = new_hidden_states

            next_pos, new_hidden_states = self.call_decoder(
                nodal_input, hidden_states, training=training
            )
            nodal_output_seq.append(next_pos)

        nodal_output_seq = tf.transpose(tf.stack(nodal_output_seq), perm=[1,0,2])  ########## shape: (TOTAL_TIMEPOINT_NO, -1, shape_dim) -> (-1, TOTAL_TIMEPOINT_NO, shape_dim)
        return self.round_differentiable(nodal_output_seq, settings['round-decimal'])



    #----------- Mean squared error (distance) in nodal trajectory
    def calculate_loss(self, batch_nodal_true, batch_nodal_pred):
        loss = tf.math.squared_difference(batch_nodal_true, batch_nodal_pred)
        return tf.reduce_mean(loss, axis=[-2,-1])
    #----------- Mean error distance in nodal trajectory
    def calculate_mean_ed(self, batch_nodal_true, batch_nodal_pred):
        acc = tf.math.squared_difference(batch_nodal_true, batch_nodal_pred)
        acc = tf.sqrt(acc)
        return tf.reduce_mean(acc, axis=[-2,-1])
    #----------- Max. error distance in nodal trajectory
    def calculate_max_ed(self, batch_nodal_true, batch_nodal_pred):
        acc = tf.math.squared_difference(batch_nodal_true, batch_nodal_pred)
        acc = tf.sqrt(acc)
        return tf.reduce_max(acc, axis=[-2,-1])

    def calculate_R2(self, batch_nodal_true, batch_nodal_pred):
      '''tf1, tf2 = tf.convert_to_tensor(batch_nodal_true), tf.convert_to_tensor(batch_nodal_pred)
      tf1 = tf.cast(tf1, tf.float32)'''
      ss_res = tf.reduce_sum(
          tf.math.squared_difference(batch_nodal_true, batch_nodal_pred), axis=-1)  #--- shape: (num_coord,)
      ss_tot = tf.reduce_sum(
          tf.math.squared_difference(batch_nodal_true, tf.reduce_mean(batch_nodal_true, keepdims=True)), axis=-1)

      r2 = tf.math.subtract(
          1., tf.math.divide(ss_res, ss_tot))
      r2 = tf.reduce_mean(r2)

      '''print('R^2 per coord: {}'.format(r2))'''
      return r2  #--- shape: (num_coord,)

    def compute_loss_and_acc(self, batch_nodal_true, batch_nodal_pred):
        tf1, tf2 = tf.convert_to_tensor(batch_nodal_true), tf.convert_to_tensor(batch_nodal_pred)
        tf1 = tf.cast(tf1, tf.float32)

        loss = self.calculate_loss(tf1, tf2)
        mean_ed = self.calculate_mean_ed(tf1, tf2)
        max_ed = self.calculate_max_ed(tf1, tf2)
        R2 = self.calculate_R2(tf1, tf2)
        return loss, mean_ed, max_ed, R2

    #@tf.function #----------- Run eagerly
    def train_step(self, batch_group):
        batch_design_idx, batch_design_input, batch_nodal_true = batch_group

        with tf.GradientTape() as tape:
            batch_nodal_pred = self.teacher_forcing_generate(batch_design_input, batch_nodal_true[:,:-1,:], training=True)
            batch_loss, batch_acc1, batch_acc2, batch_R2 = self.compute_loss_and_acc(batch_nodal_true[:,1:,:], batch_nodal_pred)

        train_vars = self.trainable_variables

        grads = tape.gradient(batch_loss, train_vars)

        self.optimizer.apply_gradients(zip(grads, train_vars))

        self.loss_tracker.update_state(batch_loss)
        self.acc_tracker1.update_state(batch_acc1)
        self.acc_tracker2.update_state(batch_acc2)
        self.R2_tracker.update_state(batch_R2)

        return {
            "loss": self.loss_tracker.result(),
            "mean_ed": self.acc_tracker1.result(),
            "max_ed": self.acc_tracker2.result(),
            "R2": self.R2_tracker.result()
        }

    #@tf.function #----------- Run eagerly
    def test_step(self, batch_group):

        batch_design_idx, batch_design_input, batch_nodal_true = batch_group

        batch_nodal_pred_infer = self.recursive_generate(batch_design_input, batch_nodal_true[:,0,:], training=False)
        batch_loss, batch_acc1, batch_acc2, batch_R2 = self.compute_loss_and_acc(batch_nodal_true[:,1:,:], batch_nodal_pred_infer)
        self.loss_tracker.update_state(batch_loss)
        self.acc_tracker1.update_state(batch_acc1)
        self.acc_tracker2.update_state(batch_acc2)
        self.R2_tracker.update_state(batch_R2)

        return {
            "loss": self.loss_tracker.result(),
            "mean_ed": self.acc_tracker1.result(),
            "max_ed": self.acc_tracker2.result(),
            "R2": self.R2_tracker.result()
        }

    def predict_step(self, batch_group):
        batch_design_idx, batch_design_input, batch_nodal_true = batch_group
        batch_nodal_pred_infer = self.recursive_generate(batch_design_input, batch_nodal_true[:,0,:], training=False)
        return batch_nodal_pred_infer

    #---------- List all metrics to reset automatically every epoch
    @property
    def metrics(self):
        return [self.loss_tracker, self.acc_tracker1, self.acc_tracker2, self.R2_tracker]
```
