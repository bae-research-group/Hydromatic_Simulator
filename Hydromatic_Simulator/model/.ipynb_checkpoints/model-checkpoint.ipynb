```python
from keras.models import import_model
from model.modelTrainer import *
from model.modelTrainer_DI import *

def load_trained_model(weights_path="model/weights", trial_id=0):

    settings['timesteps'] = ['30min', '60min', '120min', '180min', '240min', '300min', '420min', '600min', '1440min']
    settings['dataset-fname'] = '8coords-24h-dataset.csv'
    settings['num-timesteps'] = len(settings['timesteps'])
    weights_path += '/24h'
    
    modelTrainer_group_24h = load_init_model(weights_path, trial_id)
    #modelTrainer_DI_group = load_init_model_di(weights_path, trial_id)

    settings['timesteps'] = ['30min', '60min', '90min', '120min', '150min', '180min']
    settings['dataset-fname'] = '8coords-3h-dataset.csv'
    settings['num-timesteps'] = len(settings['timesteps'])
    weights_path += '/3h'
    
    modelTrainer_group_3h = load_init_model(weights_path, trial_id)
    #modelTrainer_DI_group = load_init_model_di(weights_path, trial_id)
    
    return modelTrainer_group_24h, modelTrainer_group_3h
```

```python
def load_init_model(foldername, trial_id):
  modelTrainer_group = []

  for i in range(settings['num-coord']):
      modelTrainer = ModelTrainer()
      modelTrainer_group.append(modelTrainer)

  #---------- Early-stopping condition
  lr_schedule = keras.optimizers.schedules.ExponentialDecay(
      initial_learning_rate=settings['lr'],
      decay_steps=settings['decay-step'],
      decay_rate=settings['decay-rate']
  )
  callback = keras.callbacks.EarlyStopping(
      monitor='val_loss',
      patience=5, restore_best_weights=True,
      start_from_epoch=15, mode='min'
  )


  train_history_lst = []
  for i in range(settings['num-coord']):
    if i in settings['index_lst']:
      print("\n\n###########################################")
      print("COORD {}:\n".format(str(i)))

      train_ = settings['train_lst'][i]
      val_ = settings['val_lst'][i]
      test_ = settings['test_lst'][i]


      train_dataset = make_dataset(
            settings['batch-size'], train_.idx, train_.design, train_.X, train_.Y
      )
      val_dataset = make_dataset(
            settings['batch-size'], val_.idx, val_.design, val_.X, val_.Y
      )
      test_dataset = make_dataset(
            settings['batch-size'], test_.idx, test_.design, test_.X, test_.Y
      )

      modelTrainer_group[i].compile(
          optimizer=keras.optimizers.Adam(learning_rate=lr_schedule),
          run_eagerly=True
      )

    modelTrainer_group[i].fit(
        train_dataset,
        epochs=1,
        validation_data=val_dataset,
    )
    
    modelTrainer_group[i].encoder.load_weights(
      os.path.join('.',foldername,'encoder_{}coords_{}.weights.h5'.format(settings['num-coord'], str(i))))
    modelTrainer_group[i].decoder.load_weights(
      os.path.join('.',foldername,'decoder_{}coords_{}.weights.h5'.format(settings['num-coord'], str(i))))

    #modelTrainer_group[i].evaluate(test_dataset)

  return modelTrainer_group
```

```python
def import_data():
    df = pd.read_csv(settings['dataset-fname'], dtype={'mater-dist': 'U'})[:settings['dataset-size']] #------ cut the size to settings['dataset-size']

    import_group = (
        df['design-no'].to_numpy(), df['mater-dist'].to_numpy(), df['coord-x-groundtruth'].to_numpy(), df['coord-y-groundtruth'].to_numpy())

    designs = np.array([list(x) for x in import_group[1]]).astype(float)
    X = np.array([literal_eval(x) for x in import_group[2]]).astype(float)
    Y = np.array([literal_eval(x) for x in import_group[3]]).astype(float)

    X = np.round(X, decimals=settings['round-decimal'])
    Y = np.round(Y, decimals=settings['round-decimal'])

    dataset_lst = []
    idx_lst = [0] + [idx+1 for idx, t in enumerate(settings['timesteps']) if settings['timesteps'].count(t) == 1]

    for i in range(settings['num-coord']):
        dataset = [
            import_group[0], # actuator design index
            designs, # actuator design (input to encoding layer)
            np.stack([X[:, idx, i] for idx in idx_lst], axis=1),
            np.stack([Y[:, idx, i] for idx in idx_lst], axis=1)]

        dataset_lst.append(dataset)


    settings['dataset-imported'] = dataset_lst
    return dataset_lst


def pick_rand_set(dataset_imported):

  FoldData = namedtuple("i", ["idx", "design", "X", "Y"])
  design_num, designs, Xs, Ys = dataset_imported[0], dataset_imported[1], dataset_imported[2], dataset_imported[3]

  X = np.arange(design_num.shape[0])
  train_indices, test_indices = train_test_split(
      X, test_size=settings['testset-ratio'], shuffle=True)

  test_ = FoldData(design_num[test_indices], designs[test_indices], Xs[test_indices], Ys[test_indices])


  sub_train_indices, sub_val_indices = train_test_split(
      X, test_size=settings['valset-ratio'], shuffle=True)

  train_ = FoldData(design_num[sub_train_indices], designs[sub_train_indices], Xs[sub_train_indices], Ys[sub_train_indices])

  val_ = FoldData(design_num[sub_val_indices], designs[sub_val_indices], Xs[sub_val_indices], Ys[sub_val_indices])


  return train_, val_, test_

# ----------- Import dataset -----------
def generate_dataset():
  dataset_imported = import_data()
  dataset_imported = settings['dataset-imported']

  train_lst, val_lst, test_lst = [],[],[]

  for coord_num in range(settings['num-coord']):
    train_, val_, test_ = pick_rand_set(dataset_imported[coord_num])

    train_lst.append(train_)
    val_lst.append(val_)
    test_lst.append(test_)


  settings['train_lst'] = train_lst
  settings['val_lst'] = val_lst
  settings['test_lst'] = test_lst

  return


def make_dataset(batch_size, design_idx, design_input, true_nodal_x, true_nodal_y):
    tf1 = tf.convert_to_tensor(design_idx)
    tf2 = tf.convert_to_tensor(design_input)
    tf3 = tf.stack(
        [tf.convert_to_tensor(true_nodal_x), tf.convert_to_tensor(true_nodal_y)], axis=-1)
    dataset = tf.data.Dataset.from_tensor_slices((tf1, tf2, tf3))
    # DO NOT RANDOMIZE: keep X and Y coords aligned
    return dataset.batch(batch_size).prefetch(tf.data.AUTOTUNE)
```
