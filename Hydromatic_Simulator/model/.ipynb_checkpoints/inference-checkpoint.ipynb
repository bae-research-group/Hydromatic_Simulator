```python
# model/inference.py
import numpy as np
from model.model import load_trained_model

# You can move constants like S (input length) here
S = 65

def preprocess_binary_input(binary_str):
    """Convert binary string to numpy array."""
    return np.array([int(bit) for bit in binary_str[:S]]).reshape(1, S)

def run_inference(modelTrainer_group, binary_str):
    design_input = preprocess_binary_input(binary_str)
    shape_seq_pred = predict_(modelTrainer_group, design_input)  # Assume y_pred is (1, T) or (1, T, D)
    return shape_seq_pred.squeeze()
```

```python
import io

def calculate_RMSEs(true, pred):
    mse = np.mean(np.sum((true - pred)**2, axis=-1), axis=-1)  #----- over coord (xy)
    rmse = np.sqrt(mse)  #----- shape: (timestep,)
    return np.mean(rmse), np.max(rmse)  #----- over timestep

import matplotlib.pyplot as plt
import matplotlib.ticker

'''def plot_result(true, pred, no, save_fig=False, save_name=None, fig_show=False):
    title_lst = settings['timesteps']
    num_plots = settings['num-timesteps']

    fig, axs = plt.subplots(5, 2, figsize=(10,25), constrained_layout=True)
    axs = axs.flatten()

    for t in range(num_plots):
        ax = axs[t]
        ax.set_title(title_lst[t], fontsize=40, pad=15, weight='bold')

        x_true = [0.0] + true[t, :, 0].tolist()
        y_true = [0.0] + true[t, :, 1].tolist()
        x_pred = [0.0] + pred[t, :, 0].tolist()
        y_pred = [0.0] + pred[t, :, 1].tolist()

        ax.plot(x_true, y_true, c='b', marker='o', linewidth=2, markersize=10, label='True')
        ax.plot(x_pred, y_pred, c='r', marker='o', linewidth=2, markersize=10, label='Pred')

        ax.set_xticks([0, 20, 40])
        ax.set_yticks([0, -20, -40])
        ax.xaxis.set_major_locator(matplotlib.ticker.MultipleLocator(20))
        ax.xaxis.set_minor_locator(matplotlib.ticker.MultipleLocator(10))
        ax.yaxis.set_major_locator(matplotlib.ticker.MultipleLocator(20))
        ax.yaxis.set_minor_locator(matplotlib.ticker.MultipleLocator(10))

        ax.tick_params(which='major', labelsize=35, length=8,
                       direction='in', width=1, color='k', pad=8)
        ax.tick_params(which='minor', labelsize=35, length=5,
                       direction='in', width=1, color='k', pad=8)

        ax.xaxis.set_ticks_position('both')
        ax.yaxis.set_ticks_position('both')

    # Hide unused subplots if num_plots < 10
    for i in range(num_plots, 10):
        fig.delaxes(axs[i])

    if save_fig and save_name is not None:
        save_helper(plt, save_name)

    if fig_show is True:
      plt.show()
    else:
      plt.close()'''


def plot_result(true, pred, no, save_fig=False, save_name=None, fig_show=False):
    title_lst = settings['timesteps']
    for t in range(settings['num-timesteps']):
        plt.figure(edgecolor='k', layout='constrained', frameon=False, figsize=(5,5))
        plt.rcParams['axes.linewidth'] = 1.2

        x_true = [0.0] + true[t,:,0].tolist()
        y_true = [0.0] + true[t,:,1].tolist()
        x_pred = [0.0] + pred[t,:,0].tolist()
        y_pred = [0.0] + pred[t,:,1].tolist()

        plt.plot(x_true, y_true, c='b', marker='o', linewidth=2, markersize=10, label='True')
        plt.plot(x_pred, y_pred, c='r', marker='o', linewidth=2, markersize=10, label='Pred')

        ax = plt.gca()
        plt.title(title_lst[t], fontsize=50, pad=15)

        ax.xaxis.set_ticks_position('both')
        ax.yaxis.set_ticks_position('both')
        ax.set_xticks([0, 20, 40])
        ax.set_yticks([0, -20, -40])
        ax.xaxis.set_major_locator(matplotlib.ticker.MultipleLocator(20))
        ax.xaxis.set_minor_locator(matplotlib.ticker.MultipleLocator(10))
        ax.yaxis.set_major_locator(matplotlib.ticker.MultipleLocator(20))
        ax.yaxis.set_minor_locator(matplotlib.ticker.MultipleLocator(10))

        plt.tick_params(which='major', labelsize=34, length=12, axis='both',
                        direction='in', width=1, color='k', pad=10)
        plt.tick_params(which='minor', labelsize=34, length=7, axis='both',
                        direction='in', width=1, color='k', pad=10)

        if save_fig == True:
          save_helper(plt, save_name + '_' + str(title_lst[t]))

        plt.close()
        #plt.show()



def predict_(modelTrainer_group, design_input):

  init_nodal_pos = settings['coord_8_init'] if settings['num-coord'] == 8 else settings['coord_16_init']
  nodal_pos = []

  for i, model_ in enumerate(modelTrainer_group):
    out = model_.recursive_generate(
        tf.convert_to_tensor([design_input]), tf.convert_to_tensor([init_nodal_pos[i]]), training=False)
    nodal_pos.append(out)

  return np.stack(nodal_pos, axis=1) #######--------------

#--------------------------------- Testing ------------------------------------
def test_(modelTrainer_group):

  shape_seq_true, shape_seq_pred = [], []

  for i, model_ in enumerate(modelTrainer_group):
    print(f"COORD {i}: loss/accuracy:")

    test_ = settings['test_lst'][i]
    test_dataset = make_dataset(
        settings['batch-size'], test_.idx, test_.design, test_.X, test_.Y)

    '''model.test_step = MethodType(ModelTrainer.test_step, model)'''

    model_.evaluate(test_dataset)

    out = model_.predict(test_dataset)
    shape_seq_pred.append(out)

    shape_seq_true.append(
        np.stack([test_.X[:,1:], test_.Y[:,1:]], axis=-1))

    print()

  return np.stack(shape_seq_true, axis=0), np.stack(shape_seq_pred, axis=0), np.stack(test_.idx, axis=0), np.stack(test_.design, axis=0)
```
