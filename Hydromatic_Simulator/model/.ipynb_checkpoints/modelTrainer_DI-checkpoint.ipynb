```python
class FineTuningModel(tf.keras.Model):
    def __init__(self, model_base, nodal_dist_GT, alpha,
                 model_ref=None, use_reference=False):
        super().__init__()
        self.model_base = model_base
        self.model_ref = model_ref
        self.use_reference = use_reference

        self.nodal_dist_GT = tf.convert_to_tensor(nodal_dist_GT, dtype=tf.float32)
        self.alpha = alpha

        #----- metric trackers
        self.loss_tracker = tf.keras.metrics.Mean(name="loss_w")
        self.acc_tracker1 = tf.keras.metrics.Mean(name="mean_ed")
        self.acc_tracker2 = tf.keras.metrics.Mean(name="max_ed")
        self.R2_tracker = tf.keras.metrics.Mean(name="R2")

    def compile(self, optimizer, **kwargs):
        super().compile(**kwargs)
        self.optimizer = optimizer

    def train_step(self, data):
        if self.use_reference:
            (batch_input, batch_input_ref), _ = data
            _, design_input, nodal_true = batch_input
            _, design_input_ref, nodal_true_ref = batch_input_ref
        else:
            batch_input = data
            _, design_input, nodal_true = batch_input

        design_input = tf.cast(design_input, tf.float32)
        nodal_true = tf.cast(nodal_true, tf.float32)

        with tf.GradientTape() as tape:
            if self.use_reference:
                nodal_pred_ref = self.model_ref.teacher_forcing_generate(
                    design_input, nodal_true_ref[:, :-1, :], training=False) #--- 같은 design에 대한 예측값이어야 하므로

                nodal_pred = self.model_base.teacher_forcing_generate(
                    design_input, nodal_true[:, :-1, :], training=True)

                loss = self.calculate_loss_modified(
                    nodal_true[:, 1:, :], nodal_pred, nodal_pred_ref)

            else:
                nodal_pred = self.model_base.teacher_forcing_generate(
                    design_input, nodal_true[:, :-1, :], training=True)

                loss = self.calculate_loss_modified0(
                    nodal_true[:, 1:, :], nodal_pred)

            mean_ed = self.model_base.calculate_mean_ed(nodal_true[:, 1:, :], nodal_pred)
            max_ed = self.model_base.calculate_max_ed(nodal_true[:, 1:, :], nodal_pred)
            R2 = self.model_base.calculate_R2(nodal_true[:, 1:, :], nodal_pred)

        grads = tape.gradient(loss, self.model_base.trainable_variables)
        self.optimizer.apply_gradients(zip(grads, self.model_base.trainable_variables))

        #----- update metrics
        self.loss_tracker.update_state(loss)
        self.acc_tracker1.update_state(mean_ed)
        self.acc_tracker2.update_state(max_ed)
        self.R2_tracker.update_state(R2)

        return {
            "loss_w": self.loss_tracker.result(),
            "mean_ed": self.acc_tracker1.result(),
            "max_ed": self.acc_tracker2.result(),
            "R2": self.R2_tracker.result()
        }

    def test_step(self, data):
        '''return self.train_step(data)'''
        if self.use_reference:
            (batch_input, batch_input_ref), _ = data
            _, design_input, nodal_true = batch_input
            _, design_input_ref, nodal_true_ref = batch_input_ref
        else:
            batch_input = data
            _, design_input, nodal_true = batch_input

        design_input = tf.cast(design_input, tf.float32)
        nodal_true = tf.cast(nodal_true, tf.float32)

        if self.use_reference:
            nodal_pred_ref = self.model_ref.recursive_generate(
                design_input, nodal_true_ref[:,0, :], training=False) #--- 같은 design에 대한 예측값이어야 하므로

            nodal_pred = self.model_base.recursive_generate(
                design_input, nodal_true[:,0, :], training=False)

            loss = self.calculate_loss_modified(
                nodal_true[:, 1:, :], nodal_pred, nodal_pred_ref)

        else:
            nodal_pred = self.model_base.recursive_generate(
                design_input, nodal_true[:,0, :], training=False)

            loss = self.calculate_loss_modified0(
                nodal_true[:, 1:, :], nodal_pred)

        mean_ed = self.model_base.calculate_mean_ed(nodal_true[:, 1:, :], nodal_pred)
        max_ed = self.model_base.calculate_max_ed(nodal_true[:, 1:, :], nodal_pred)
        R2 = self.model_base.calculate_R2(nodal_true[:, 1:, :], nodal_pred)

        #----- update metrics
        self.loss_tracker.update_state(loss)
        self.acc_tracker1.update_state(mean_ed)
        self.acc_tracker2.update_state(max_ed)
        self.R2_tracker.update_state(R2)

        return {
            "loss_w": self.loss_tracker.result(),
            "mean_ed": self.acc_tracker1.result(),
            "max_ed": self.acc_tracker2.result(),
            "R2": self.R2_tracker.result()
        }


    @property
    def metrics(self):
        return [self.loss_tracker, self.acc_tracker1, self.acc_tracker2, self.R2_tracker]

    def calculate_loss_modified0(self, y_true, y_pred):
        base_loss = tf.reduce_mean(tf.square(y_true - y_pred), axis=[-2, -1])
        penalty = tf.reduce_mean(tf.square(y_pred - self.nodal_dist_GT))
        return base_loss + self.alpha * penalty

    def calculate_loss_modified(self, y_true, y_pred, y_pred_ref):
        base_loss = tf.reduce_mean(tf.square(y_true - y_pred), axis=[-2, -1])
        nodal_diff = y_pred - y_pred_ref
        penalty = tf.reduce_mean(tf.square(nodal_diff - self.nodal_dist_GT))
        return base_loss + self.alpha * penalty
```
