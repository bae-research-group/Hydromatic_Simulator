```python
# utils/plotting.py
import matplotlib.pyplot as plt
import os

def save_sequence_plot(pred_sequence, output_dir="outputs"):
    os.makedirs(output_dir, exist_ok=True)
    
    # Assume pred_sequence is (T,) or (T, D)
    num_frames = len(pred_sequence)
    image_paths = []
    
    for i, y in enumerate(pred_sequence):
        fig, ax = plt.subplots()
        ax.plot(y if hasattr(y, "__len__") else [y])  # handle 1D or 2D
        ax.set_title(f"Prediction Frame {i}")
        filename = os.path.join(output_dir, f"frame_{i}.png")
        fig.savefig(filename)
        image_paths.append(filename)
        plt.close(fig)

    return image_paths
```
